# Analysis

Research question: Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

Hint:

To compute usage, we will need to understand that we have to join successive (in time) measures/counters for a given station, as the difference will tell whether a car was picked up, returned, or nothing happened.

```python
# Step 1

# From the hint, let's find the difference in the number of bluecars at the station 
# by creating a new column Bluecar_difference

new_df['Bluecar_difference'] = new_df['Bluecar counter'].diff()
new_df.head()
```

Interpretation of the difference column:
1. Positive value means a car has been returned
2. Negative value means a car has been picked 
3. 0 means nothing has happened

```python
# Step 2

# Selecting the city of Paris only and filtering rental status to be operational

paris = new_df[(new_df['City'] == 'Paris') & (new_df['Rental status'] == 'operational')]
paris.head()
```

```python
# Step 3
# Getting the most popular hour for picking up Bluecars

# creating a pivot table to display the cars and the hours
p_table = pd.pivot_table(paris, values = ['Bluecar counter', 'Utilib counter', 'Utilib 1.4 counter'], index = ['hour'], aggfunc = np.sum)
p_table

# find the total number of cars by hour
cars_total = p_table['Bluecar counter'] + p_table['Utilib 1.4 counter'] + p_table['Utilib counter']
p_table['Total'] = cars_total

# sort the total number of cars in descending order to find the most popular picking hour
p_table.sort_values(by = 'Total', ascending = False).head(1)

```

```python
# most popular hour for returning cars 

# using slots: the bigger the number of slots the more cars being rented, the smaller the number of slots the more cars being returned

# creating a pivot table for displaying the sum of slots available per hour 
returning = pd.pivot_table(paris, values = ['Slots'], index = ['hour'], aggfunc = np.sum)

# sorting the slots in ascending order 
returning.sort_values(by = 'Slots').head(1)

```

```python
# most popular station overally

# finding sum of all cars in a station 
paris['cars'] = paris['Bluecar counter'] + paris['Utilib 1.4 counter'] + paris['Utilib counter']

# sorting the cars in descending order to get most popular station and displaying using a pivot table 
paris.pivot_table(values = ['cars'], index = ['ID'], aggfunc = np.sum).sort_values(by = 'cars', ascending = False).head(5)

```

```python
# most popular station at the most popular picking hour (hour 21)


popular_station = paris[paris['hour'] == 21]
station_name = pd.pivot_table(popular_station, values = ['Bluecar counter', 'Utilib 1.4 counter', 'Utilib counter'], index = ['ID', 'Station type'], aggfunc = np.sum) 
summation = station_name['Bluecar counter'] + station_name['Utilib 1.4 counter'] + station_name['Utilib counter']
station_name['sum'] = summation
station_name.sort_values(by = 'sum', ascending = False).head(1)
```

```python
# What postal code is the most popular for picking up Blue cars? 
# Does the most popular station belong to that postal code? No

paris.pivot_table(values = 'Bluecar counter', index = ['Postal code', 'ID'], aggfunc = np.sum).sort_values(by = 'Bluecar counter', ascending = False).head(2)

```

```python
# Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars? 

utilibs = paris['Utilib 1.4 counter'] + paris['Utilib counter']
paris['total utilibs'] = utilibs
paris.pivot_table(values = ['total utilibs'], index = ['hour', 'Postal code', 'ID'], aggfunc = np.sum).sort_values(by = 'total utilibs', ascending = False)


# most popular picking hour = 14
# most popular postal code = 75001
# most popular station is paris-jeanjacquesrousseau-62
```
